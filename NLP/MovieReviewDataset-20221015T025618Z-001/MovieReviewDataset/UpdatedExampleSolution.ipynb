```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sentimentAnalyser = SentimentIntensityAnalyzer()
```

```python
sentimentAnalyser.polarity_scores("I am happy")
```

```python
sentimentOutput = sentimentAnalyser.polarity_scores("I am happy")
sentimentOutput.pop('compound')
```

```python
sorted(sentimentOutput.items() , key = lambda x:x[1], reverse=True)[0][0]
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def identifySentiment(document):
  sentimentAnalyser = SentimentIntensityAnalyzer()
  sentimentOutput = sentimentAnalyser.polarity_scores(document)
  sentimentOutput.pop('compound')
  result = sorted(sentimentOutput.items() , key = lambda x:x[1], reverse=True)[0][0] 
  if result == 'neu':
    return 'pos'
  else:
    return result
```

```python
identifySentiment("I am Happy")
```

```python
#Use apply()
data['predLabel'] = data['review'].apply(identifySentiment)
data.head()
```

```python
data.predLabel.value_counts()
```

```python
identifySentiment(data['review'][1])
```

```python
# https://blog.quantinsti.com/vader-sentiment/

# Compound Score ---> -1 to 1
# if compound score > 0.5 ---> pos
# else ----------------------> neg
```

```python
data['myCompoundLabel'] = data['review'].apply(lambda record: ['pos' if sentimentAnalyser.polarity_scores(record)['compound'] > 0.5 else 'neg'][0])
data.head()
```

```python
data['myCompoundLabel'].value_counts()
```

```python
#Checking Accuracy
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(data['label'],data['predLabel'] ))
print(accuracy_score(data['label'],data['myCompoundLabel'] ))
```
